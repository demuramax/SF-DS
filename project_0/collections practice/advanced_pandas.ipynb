```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python

melb_df['Date'] = pd.to_datetime(melb_df['Date'])
# print(melb_df['Date'])
quarter_sold = melb_df['Date'].dt.quarter
quarter_sold.value_counts().iloc[1]
```

```python

melb_df.columns

unique_list = []

for col in melb_df.columns:
    item = (col, melb_df[col].nunique(), melb_df[col].dtype)
    unique_list.append(item)
    
unique_counts = pd.DataFrame(
    unique_list,
    columns = ['Column Name', 'Num Unique', 'Type']).sort_values(by = 'Num Unique', ignore_index=True)

print(unique_counts)
```

```python
black_list = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_counts = 150

for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_counts and col not in black_list:
        melb_df[col] = melb_df[col].astype('category')
        

melb_df.info()
print(melb_df.dtypes.count())
```

```python
melb_df.sort_values(by='Date', ascending=False)
melb_df.sort_values(by=['Price', 'Distance']).loc[::10, ['Distance', 'Price']]
```

```python
mask1 = melb_df['AreaRatio'] <= -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'

melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
```

```python
melb_df.sort_values(by=['AreaRatio'], ascending=False, ignore_index=True).iloc[1558]
```

```python
melb_df[(melb_df['Type'] == 'townhouse') & (melb_df['Rooms'] > 2)].sort_values(by=['Rooms', 'MeanRoomsSquare'], ascending=[True, False], ignore_index=True).iloc[18]

```

```python
melb_df.groupby(by='Type')['Price'].mean()
melb_df.groupby(by='Regionname')['Distance'].min().sort_values(ascending=False)
melb_df.groupby(by='MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False)

melb_df.groupby(by='Regionname')['SellerG'].agg(['nunique', set])
```

```python
melb_df.groupby(by='Rooms')['Price'].mean().sort_values(ascending=False)
melb_df.groupby(by='Regionname')['Lattitude'].min().sort_values()

melb_df['Date']
date1 = pd.to_datetime('2017-05-01')
date2 = pd.to_datetime('2017-10-01')
dates_range = (melb_df['Date'] >= date1) & (melb_df['Date'] < date2)

melb_df[dates_range].groupby(by='SellerG')['Price'].sum().sort_values(ascending=True)
```

```python
melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack()

melb_df.pivot_table(
    values = 'Price',
    index = 'Rooms',
    columns = 'Type'
).round()
```

```python
melb_df.pivot_table(
    values = 'Price',
    index = 'Regionname', 
    columns = 'Weekend',
    aggfunc='count'
)

melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
    
).round()

melb_df.pivot_table(
    values='Price',
    index=['Method', 'Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
).round()
```

```python
pivot = melb_df.pivot_table(
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)

pivot.columns
pivot['mean']['unit']

mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
filtered_pivot
print(list(filtered_pivot.index))
```
