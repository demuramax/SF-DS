```python
import pandas as pd

ratings1 = pd.read_csv('data/movies_data/ratings1.csv')
ratings2 = pd.read_csv('data/movies_data/ratings2.csv')
movies = pd.read_csv('data/movies_data/movies.csv')
dates = pd.read_csv('data/movies_data/dates.csv')
```

```python
dates['date'] = pd.to_datetime(dates['date'])
years = dates['date'].dt.year
years.mode()
```

```python
ratings = pd.concat(
    [ratings1, ratings2], 
    ignore_index=True
)
ratings

print('Number of rows in table Ratings:', ratings.shape[0])
print('Number of rows in table dates:', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])

print(ratings1.tail(1))
print(ratings2.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Number of rows in the table Ratings:', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
ratings_dates.tail(7)
movies
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
joined

merged = ratings_dates.merge(
    movies,
    on='movieId', 
    how='left'
)
merged
```
